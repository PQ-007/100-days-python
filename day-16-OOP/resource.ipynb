### Why we need OOP (Object Oriented Programming)?

When using Procedural programming it become hard to maintain our project, code base. One hot block of code that contain every function of our project is like superman that can handle everything. To scale up, to understand etc, it becomes really difficult. Then we need OOP. Object Oriented Programming aka OOP is programming paradigm, that can separate project identidies by making reusable pieces. Making smaller component component named object is the main idea. Object contains methods and attribute.
>Object has methods and attributes. Method is what it can do, attribute is what it has.

For example waiter can: Take orders, Take payments, Serve dishes
waiter has: dishes, money, order etc
