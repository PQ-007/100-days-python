```python
print("Welcome to the rollercoaster!")
height = int(input("What is your height in cm? "))

if height > 120:
    print("You can ride rollercoaster.")
else:
    print("Please go away, Go home.")
```

```python
#Modulo
number = int(input("Enter a number: "))
if number % 2 == 0:
    print("This is an even number.")
else:    print("This is an odd number.")
```

```python
#Nested if statements
print("Welcome to the rollercoaster!")
height = int(input("What is your height in cm? "))
if height > 120:
    print("You can ride rollercoaster.")
    age = int(input("What is your age? "))
    if age < 12:
        print("Please pay $5.")
    elif age <= 18:
        print("Please pay $7.")
    else:
        print("Please pay $12.")
else:
    print("Please go away, Go home.")
```

```python
#Exercise
weight = 85
height = 1.85

bmi = weight / (height ** 2)

if bmi < 18.5:
    print("underweight")
elif bmi >= 18.5 and bmi < 25:
    print("normal weight")
elif bmi <= 25:
    print("overweight")
```
