```python
#loops

#for loop
fruits = ["Apple", "Banana", "Cherry", "Date", "Elderberry", "Fig", "Grape", "Honeydew", "Kiwi", "Lemon", "Mango", "Nectarine"]
for item in fruits:
    print(item)
for number in range(1, 10):
    print(number)

#while loop
count = 0
while count < 5:
    print(count)
    count += 1

#do-while loop
while True:
    user_input = input("Type 'exit' to quit: ")
    if user_input.lower() == 'exit':
        break
```

```python
#break statement exits the loop immediately
for number in range(1, 10):
    if number == 5:
        break
    print(number)

#continue statement breaks the current iteration and moves to the next one
for number in range(1, 10):
    if number % 2 == 0:
        continue
    print(number)
    
#pass statement is a placeholder that does nothing
for number in range(1, 10):
    if number % 2 == 0:
        pass
    else:
        print(number)
```

```python
#highest score challenge
student_scores = [78, 65, 89, 86, 55, 91, 64, 89]
highest_score = 0
for score in student_scores:
    if score > highest_score:
        highest_score = score
print(highest_score)
print(max(student_scores))
total_score = sum(student_scores)
print(total_score)
average_score = total_score / len(student_scores)
print(average_score)
```
