# Python-ын үндсэн өгөгдлийн төрөлүүд

| Төрөл | Жишээ | Тайлбар |
|---|---:|---|
| int (Бүхэл тоо) | 42 | Сөрөг/эерэг бүхэл тоонууд |
| float (Бутархай/хөвөгч цэгийн тоо) | 3.14 | Бутархай утгатай тоонууд |
| bool (Логик) | True, False | Жинхэнэ/худал (мөрт) үнэн шалгалт |
| str (Мөр/стринг) | "Сайн уу" | Тэмдэгт мөрүүд |
| list (Жагсаалт) | [1, 2, 3] | Өөрчлөгдөх дараалсан контейнер |
| tuple (Тогтмол жагсаалт) | (1, 2) | Өөрчлөгдөхгүй дараалал |
| set (Цуглуулга) | {1, 2, 3} | Давхцалгүй элемэнтүүдийн эзлэхүүн |
| dict (Түлхүүр-утга) | {'a': 1} | Түлхүүрээр индексжүүлэх хосууд |
| NoneType (None) | None | Утгагүй эсвэл тодорхой бус төлөв |
| bytes (Бинар өгөгдөл) | b'abc' | Байтов массив, бинар өгөгдөл |


```python
#type checker
type(5)
type(5.5)
type("hello")
type(True)

#convertion
print(float(5))
print(int(5.5))
print(str(5))
print(bool(5))
print(bool(0))
```

```python
#math operations
print(5 + 3)
print(5 - 3)
print(5 * 3)
print(5 / 3)
print(5 // 3)
print(5 % 3)
print(5 ** 3)
```

```python
#bmi calculator
height = input("enter your height in m: ")
if int(height) > 100:
    height = int(height) / 100
weight = input("enter your weight in kg: ") 
bmi = int(weight) / float(height) ** 2
print(int(bmi))
```

```python
#tip calculator
print("Welcome to the tip calculator!")
bill = float(input("What was the total bill? $"))
tip = int(input("How much tip would you like to give? 10, 12, or 15? "))
people = int(input("How many people to split the bill? "))
tip_as_percent = tip / 100
total_tip_amount = bill * tip_as_percent
total_bill = bill + total_tip_amount
bill_per_person = total_bill / people
final_amount = round(bill_per_person, 2)
final_amount = "{:.2f}".format(bill_per_person)
print(f"Each person should pay: ${final_amount}")
```
